```python
import numpy as np
import matplotlib.pyplot as plt

sec = 48  # 信号の長さ s
sf = 24 # サンプリング周波数 Hz

a= -315 #　振幅
c=-10 #　頂点位相
d=0 #　Mesor

e= -315 #　振幅
g=-90 #　頂点位相
h=0 #　Mesor

i= -315 #　振幅
j=-180 #　頂点位相
k=0 #　Mesor

l= -315 #　振幅
m=-270 #　頂点位相
n=0 #　Mesor

t = np.arange(0, sec, 1/sf) #サンプリング点の生成

y = a*np.cos(6.2832*t/24+c)+d # コサイナー1の式

y_ = e*np.cos(6.2832*t/24+g)+h # コサイナー2の式

y__= i*np.cos(6.2832*t/24+j)+k # コサイナー3の式

y___= l*np.cos(6.2832*t/24+m)+n # コサイナー4の式


#1つ目のデータをplotメソッドでプロット
plt.plot(t, y, color = 'red', marker = 'o')

#2つ目のデータをplotメソッドでプロット
plt.plot(t, y_, color = 'blue', marker = 'v')

#3つ目のデータをplotメソッドでプロット
plt.plot(t, y__, color = 'yellow', marker = 'v')

#4つ目のデータをplotメソッドでプロット
plt.plot(t, y___, color = 'green', marker = 'v')


```
