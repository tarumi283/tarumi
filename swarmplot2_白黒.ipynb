seaborn.swarmplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, **kwargs)

#csvデータからswarm plot(beeswarm plot)を作成。

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/demo_.csv')
data.head()
ax = sns.swarmplot(x=data["group"], y=data["value"])

means = data.groupby('group').mean()['value']
errors = data.groupby('group').std()['value'] / np.sqrt(data.groupby('group').count()['value'])
ax = means.plot.bar(yerr=errors,figsize=(5,3))
```

#swarmplot + boxplot

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
data = pd.read_csv('/content/result.csv', encoding="shift-jis")
data.head()
ax = sns.swarmplot(x=data["group"], y=data["value"],color=".2", edgecolor="white")
ax=sns.set_palette('Set3')

ax=sns.boxplot(x=data["group"], y=data["value"],width=0.6,order=data.group.unique(),whis=np.inf)
ax.set_ylabel("Period(hour)")
ax.set_xlabel("")
ax.set_yticks([22.5,23,23.5])

plt.savefig("file_name.jpeg",dpi=500,bbox_inches="tight")
files.download("file_name.jpeg")
```

#strips + box plot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files

# Assuming 'result.csv' file is present in the specified path
tips = pd.read_csv('/result_damping rate.csv', encoding="shift-jis")

# Set colors to black and white
box_line_color = "black"
box_fill_color = "white"
stripplot_color = "black"

# Create boxplot with specified box line and fill colors
ax = sns.boxplot(x="group", y="value", data=tips, whis=np.inf, color=box_fill_color,
                 boxprops=dict(edgecolor=box_line_color, facecolor=box_fill_color))

# Add stripplot on top with black points
ax = sns.stripplot(x="group", y="value", data=tips, color=stripplot_color)

ax.set_ylabel("Damping rate")
ax.set_xlabel("")
plt.xticks(rotation=90)

# Save the plot as a JPEG file
plt.savefig("file_name.jpeg",dpi=500,bbox_inches="tight")
# Download the saved file
files.download("file_name.jpeg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files

# Assuming 'result.csv' file is present in the specified path
tips = pd.read_csv('/content/result.csv', encoding="shift-jis")

# Set custom colors for the boxplot and stripplot
boxplot_color = "white"  # Change this to your desired boxplot color
stripplot_color = "white"  # Change this to your desired stripplot color
line_color = "black"  # Change this to your desired line color

# Create boxplot with white box color and black lines
ax = sns.boxplot(x="group", y="value", data=tips, whis=np.inf, color=boxplot_color, boxprops=dict(edgecolor=line_color))

# Add stripplot on top with white color and black points
ax = sns.stripplot(x="group", y="value", data=tips, color=stripplot_color, edgecolor=line_color)

ax.set_ylabel("Period(hour)")
ax.set_xlabel("")
plt.xticks(rotation=90)

# Save the plot as a JPEG file
plt.savefig("file_name.jpeg", dpi=500)

# Download the saved file
files.download("file_name.jpeg")
```
