```python
import yfinance as yf
import matplotlib.pyplot as plt

# Ticker symbols for stock indices
symbols = {
    "Nikkei 225": "^N225",  # Nikkei 225
    "S&P 500": "^GSPC",     # S&P 500
    "NASDAQ": "^IXIC",      # NASDAQ
    "Dow Jones": "^DJI"     # Dow Jones
}

# Date range for data retrieval
start_date = "1950-01-01"  # Start date
end_date = "2024-12-31"    # End date

# Retrieve and plot stock data
plt.figure(figsize=(12, 8))

for index_name, symbol in symbols.items():
    # Retrieve data using yfinance
    data = yf.download(symbol, start=start_date, end=end_date)
    if not data.empty:
        plt.plot(data["Close"], label=index_name)  # Plot closing price
    else:
        print(f"Failed to retrieve data for: {index_name}")

# Graph settings
plt.title("Stock Index Trends", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Stock Price (Close)", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()

# Display the graph
plt.show()
```
