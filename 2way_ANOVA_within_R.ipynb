```python
%%R

# 必要なパッケージのインストール
if(!require(afex)) install.packages("afex", repos='http://cran.rstudio.com/')
if(!require(dplyr)) install.packages("dplyr", repos='http://cran.rstudio.com/')
if(!require(tidyr)) install.packages("tidyr", repos='http://cran.rstudio.com/')
if(!require(effectsize)) install.packages("effectsize", repos='http://cran.rstudio.com/')

# パッケージの読み込み
library(afex)
library(dplyr)
library(tidyr)
library(effectsize)  # 効果量の計算に使用

# CSVファイルの読み込み（適宜ファイルパスを変更してください）
data <- read.csv("/content/result_exp1_within.csv")

# データの確認
head(data)

# 被験者内2要因分散分析の実行
anova_result <- aov_ez(
  id = "subject",   # 被験者IDを示す列名
  dv = "value",     # 従属変数を示す列名
  within = c("factorA", "factorB"),  # 被験者内要因を示す列名
  data = data       # データフレーム
)

# 分散分析の結果を表示
print(summary(anova_result))

# 効果量 (Partial eta squared) の計算
eta_sq_result <- eta_squared(anova_result, partial = TRUE)
print(eta_sq_result)
```

```python
%%R
# 因子Aと因子Bと被験者IDの名前を適切に設定
factorA <- 'factorA'  # 被験者内要因A
factorB <- 'factorB'  # 被験者内要因B
subject_column <- 'subject'  # 被験者ID
value_column <- 'value'  # 測定値

# 被験者内要因に対する2要因分散分析の実行
anova_result <- aov_ez(id = subject_column,
                       dv = value_column,
                       within = c(factorA, factorB),
                       data = data)

# 事後解析 (多重比較) をペアごとにt検定で実行し、Bonferroni補正を使用
# 因子Aに対するペアごとのt検定
pairwise_A <- pairwise.t.test(data[[value_column]], data[[factorA]], p.adjust.method = "bonferroni")
print("因子Aのペアごとのt検定 (Bonferroni補正):")
print(pairwise_A)

# 因子Bに対するペアごとのt検定
pairwise_B <- pairwise.t.test(data[[value_column]], data[[factorB]], p.adjust.method = "bonferroni")
print("因子Bのペアごとのt検定 (Bonferroni補正):")
print(pairwise_B)

# 因子Aと因子Bの交互作用効果に対するペアごとのt検定
# 新しい列を作成して交互作用効果を評価
data$interaction <- interaction(data[[factorA]], data[[factorB]])

pairwise_interaction <- pairwise.t.test(data[[value_column]], data$interaction, p.adjust.method = "bonferroni")
print("交互作用のペアごとのt検定 (Bonferroni補正):")
print(pairwise_interaction)
```
