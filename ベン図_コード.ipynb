```python
# 必要なライブラリをインストール
# pip install pandas

import pandas as pd

# エクセルファイルの読み込み
df = pd.read_excel('your_excel_file.xlsx')  # エクセルファイルのパスを指定

# エクセルの1列目をリストに変換
column_data = df.iloc[:, 0].tolist()  # 1列目をリストに変換

print(column_data)  # リストを表示

# 必要なライブラリをインストール
# pip install matplotlib

import matplotlib.pyplot as plt
from matplotlib_venn import venn2  # venn2関数を使用するためにimport

# データのセットアップ
set1 = set([1, 2, 3, 4, 5])
set2 = set([4, 5, 6, 7])

# ベン図を描画
venn2([set1, set2], ('Set1', 'Set2'))

# ベン図にラベルを追加
plt.annotate('Label1', xy=(-0.5, 1.5), fontsize=12)
plt.annotate('Label2', xy=(2.2, 1.5), fontsize=12)

# グラフの表示
plt.show()
```
